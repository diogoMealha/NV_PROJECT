```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
without_mag = np.load('images/no_mag.npz', allow_pickle=True)['arr_0']
with_mag = np.load('images/with_mag.npz', allow_pickle=True)['arr_0']
```

```python
with_mag
```

```python
diff =  without_mag - with_mag 
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(with_mag, cmap='gray')
plt.title('With Magnet')  
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(without_mag, cmap='gray')
plt.title('Without Magnet')  
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(diff, cmap='gray')
plt.title('Difference (Without - With)') 
```
